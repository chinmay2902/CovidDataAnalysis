```python
covid_data_max = data['Confirmed'].max()
# covid_data_max = covid_data_max.item()
world_geo = "world_countries.json"
world_map = folium.Map(location=[4.68, 8.33],titles='Mapbox Bright', zoom_start=3)
world_map = folium.Choropleth(
    geo_data=world_geo,
#     name='choropleth',
    data=data,
    columns=['Country','Confirmed'],
    key_on='feature.properties.name',
    threshold_scale = [0,int((covid_data_max/15)),int((covid_data_max/10)),int((covid_data_max/4)),covid_data_max],
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Number of Confirmed cases per country',
    highlight=True,
    line_color='black'
    ).add_to(world_map)
folium.LayerControl().add_to(world_map)
world_map.save('map.html')
```
